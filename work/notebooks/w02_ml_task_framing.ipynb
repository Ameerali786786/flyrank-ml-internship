### 1) My lane as an ML task (type)
This task is framed as a **Ranking and Scoring** problem (or Classification depending on the exact slice). We want to predict the relevance or score of a specific flight option so we can order them optimally for the user. Instead of just identifying if a flight is delayed, we score it to present the best options first.

### 2) Target or proxy
* **Target:** `flight_utility_score` or `is_preferred` (A continuous score or binary indicator representing whether a user would prefer or select this flight based on price, duration, and historical delay).
* **Proxy:** If direct user satisfaction isn't available, we use a proxy like `click_through_rate` or a composite formula of `(min_price / flight_price) - (delay_minutes / 60)`.

### 3) Success metric
* **Offline Metric (Technical):** NDCG (Normalized Discounted Cumulative Gain) for ranking accuracy, or RMSE (Root Mean Squared Error) for the continuous score prediction.
* **Online Metric (Business):** Increase in user booking conversion rate and a reduction in the time users spend searching before selecting a flight.

* **Unit of Analysis:** One row = One specific flight option offered inside a user's search query session.

### 5) Why ML beats a fixed rule here
A fixed rule like *"Always show the cheapest flight first"* breaks down quickly because users care about multiple non-linear interactions (e.g., a flight might be $10 cheaper but has a 5-hour layover or a 40% historical delay rate). Writing manual `if/else` rules for balancing price, layovers, airline reputation, and delays becomes impossible to maintain. ML dynamically learns these weights from real data.
